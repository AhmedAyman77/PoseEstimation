```python
import cv2
import time
from Modules.PoseEstimationModule import poseDetector
```

```python
cap = cv2.VideoCapture(0)
pTime = 0
detector = poseDetector()

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (820, 740))

    if not ret:
        break

    frame = detector.findPose(frame)
    lmList = detector.findPosition(frame)

    if len(lmList) != 0:
        print(lmList[20])
        cv2.circle(frame, (lmList[20][1], lmList[20][2]), 15, (0, 0, 255), cv2.FILLED)

        angle = detector.findAngle(frame, 12, 14, 16)
    
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    cv2.putText(frame, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3,
                (255, 0, 0), 3)

    cv2.imshow("Image", frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
